```python
!pip install rarfile

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_fscore_support
from sklearn.utils.class_weight import compute_class_weight
import warnings
warnings.filterwarnings('ignore')

# Deep Learning Libraries
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models, optimizers, callbacks
from tensorflow.keras.applications import VGG19
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import to_categorical
import cv2
from collections import Counter
import rarfile
import zipfile
from google.colab import files

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Function to extract RAR file
def extract_dataset():
    """Extract the uploaded RAR file"""
    print("Please upload your DataSet.rar file:")
    uploaded = files.upload()

    # Get the uploaded file name
    rar_filename = list(uploaded.keys())[0]
    print(f"Uploaded file: {rar_filename}")

    # Extract the RAR file
    if rar_filename.endswith('.rar'):
        with rarfile.RarFile(rar_filename) as rf:
            rf.extractall('.')
        print("RAR file extracted successfully!")
    elif rar_filename.endswith('.zip'):
        with zipfile.ZipFile(rar_filename, 'r') as zf:
            zf.extractall('.')
        print("ZIP file extracted successfully!")

    # Check if DataSet folder exists
    if os.path.exists('DataSet'):
        print("DataSet folder found!")
        print("Contents:", os.listdir('DataSet'))
        return 'DataSet'
    else:
        # Look for alternative folder names
        folders = [f for f in os.listdir('.') if os.path.isdir(f) and any(
            subfolder in os.listdir(f) for subfolder in ['Data_cloudy', 'Data_desert', 'Data_green_area', 'Data_water']
            if os.path.exists(f)
        )]
        if folders:
            data_path = folders[0]
            print(f"Dataset found in: {data_path}")
            return data_path
        else:
            raise FileNotFoundError("Could not find dataset with expected structure")

class ImageClassificationSystem:
    def __init__(self, data_path, img_size=(224, 224), batch_size=32):
        self.data_path = data_path
        self.img_size = img_size
        self.batch_size = batch_size
        self.num_classes = 4
        self.class_names = ['Data_cloudy', 'Data_desert', 'Data_green_area', 'Data_water']
        self.models = {}
        self.histories = {}

    def load_and_preprocess_data(self):
        """Load and preprocess the dataset"""
        print("Loading and preprocessing data...")

        # Check available classes
        available_classes = [d for d in os.listdir(self.data_path)
                           if os.path.isdir(os.path.join(self.data_path, d))]
        print(f"Available classes: {available_classes}")

        # Update class names based on available folders
        self.class_names = available_classes
        self.num_classes = len(self.class_names)

        # Count images in each class
        for class_name in self.class_names:
            class_path = os.path.join(self.data_path, class_name)
            image_count = len([f for f in os.listdir(class_path)
                             if f.lower().endswith(('.png', '.jpg', '.jpeg'))])
            print(f"{class_name}: {image_count} images")

        # Data generators with augmentation for training
        train_datagen = ImageDataGenerator(
            rescale=1./255,
            rotation_range=20,
            width_shift_range=0.2,
            height_shift_range=0.2,
            horizontal_flip=True,
            zoom_range=0.2,
            fill_mode='nearest',
            validation_split=0.3  # 70% train, 30% for val+test
        )

        # No augmentation for validation and test
        val_test_datagen = ImageDataGenerator(
            rescale=1./255,
            validation_split=0.5  # Split val+test into 15% val, 15% test
        )

        # Training data (70%)
        self.train_generator = train_datagen.flow_from_directory(
            self.data_path,
            target_size=self.img_size,
            batch_size=self.batch_size,
            class_mode='categorical',
            subset='training',
            shuffle=True,
            seed=42
        )

        # Validation data (15%)
        self.val_generator = train_datagen.flow_from_directory(
            self.data_path,
            target_size=self.img_size,
            batch_size=self.batch_size,
            class_mode='categorical',
            subset='validation',
            shuffle=False,
            seed=42
        )

        # Test data (15%) - using a separate generator
        self.test_generator = val_test_datagen.flow_from_directory(
            self.data_path,
            target_size=self.img_size,
            batch_size=self.batch_size,
            class_mode='categorical',
            subset='training',  # This will be our test set
            shuffle=False,
            seed=123  # Different seed for test
        )

        print(f"Training samples: {self.train_generator.samples}")
        print(f"Validation samples: {self.val_generator.samples}")
        print(f"Test samples: {self.test_generator.samples}")
        print(f"Classes: {list(self.train_generator.class_indices.keys())}")

        # Calculate class weights for imbalanced data
        class_counts = Counter(self.train_generator.classes)
        total_samples = sum(class_counts.values())
        self.class_weights = {i: total_samples / (len(class_counts) * count)
                            for i, count in class_counts.items()}

        print(f"Class distribution: {class_counts}")
        print(f"Class weights: {self.class_weights}")

    def build_custom_cnn(self):
        """Build custom CNN architecture as specified"""
        print("Building Custom CNN...")

        model = models.Sequential([
            # Input layer
            layers.Input(shape=(*self.img_size, 3)),

            # Convolutional Block 1
            layers.Conv2D(64, (3, 3), padding='same', name='conv1'),
            layers.BatchNormalization(name='bn1'),
            layers.ReLU(name='relu1'),
            layers.MaxPooling2D((2, 2), name='pool1'),
            layers.Dropout(0.25, name='dropout1'),

            # Convolutional Block 2
            layers.Conv2D(128, (3, 3), padding='same', name='conv2'),
            layers.BatchNormalization(name='bn2'),
            layers.ReLU(name='relu2'),
            layers.MaxPooling2D((2, 2), name='pool2'),
            layers.Dropout(0.25, name='dropout2'),

            # Convolutional Block 3
            layers.Conv2D(256, (3, 3), padding='same', name='conv3'),
            layers.BatchNormalization(name='bn3'),
            layers.ReLU(name='relu3'),
            layers.GlobalAveragePooling2D(name='global_pool'),
            layers.Dropout(0.5, name='dropout3'),

            # Classification Block
            layers.Dense(512, activation='relu', name='fc1'),
            layers.Dropout(0.5, name='dropout4'),
            layers.Dense(256, activation='relu', name='fc2'),
            layers.Dense(self.num_classes, activation='softmax', name='output')
        ])

        return model

    def build_vgg19_transfer(self):
        """Build VGG19 with transfer learning"""
        print("Building VGG19 Transfer Learning Model...")

        # Load pre-trained VGG19
        base_model = VGG19(
            weights='imagenet',
            include_top=False,
            input_shape=(*self.img_size, 3)
        )

        # Freeze base model layers initially
        base_model.trainable = False

        # Add custom classification head
        model = models.Sequential([
            base_model,
            layers.GlobalAveragePooling2D(),
            layers.Dropout(0.5),
            layers.Dense(512, activation='relu'),
            layers.BatchNormalization(),
            layers.Dropout(0.5),
            layers.Dense(256, activation='relu'),
            layers.Dropout(0.3),
            layers.Dense(self.num_classes, activation='softmax')
        ])

        return model

    def build_convnext_model(self):
        """Build ConvNeXt-like architecture (simplified version)"""
        print("Building ConvNeXt-inspired Model...")

        def conv_block(x, filters, kernel_size=7):
            # Depthwise convolution
            x = layers.DepthwiseConv2D(kernel_size, padding='same')(x)
            x = layers.BatchNormalization()(x)

            # Pointwise convolution (1x1)
            x = layers.Conv2D(filters, 1, padding='same')(x)
            x = layers.Activation('gelu')(x)

            # Another pointwise convolution
            x = layers.Conv2D(filters, 1, padding='same')(x)

            return x

        inputs = layers.Input(shape=(*self.img_size, 3))

        # Stem
        x = layers.Conv2D(96, 4, strides=4, padding='same')(inputs)
        x = layers.BatchNormalization()(x)

        # Stage 1
        x = conv_block(x, 96)
        x = layers.MaxPooling2D(2)(x)

        # Stage 2
        x = conv_block(x, 192)
        x = layers.MaxPooling2D(2)(x)

        # Stage 3
        x = conv_block(x, 384)
        x = layers.GlobalAveragePooling2D()(x)

        # Classification head
        x = layers.Dropout(0.5)(x)
        x = layers.Dense(512, activation='gelu')(x)
        x = layers.Dropout(0.3)(x)
        outputs = layers.Dense(self.num_classes, activation='softmax')(x)

        model = models.Model(inputs, outputs)
        return model

    def build_swin_transformer_like(self):
        """Build Swin Transformer-inspired model (simplified)"""
        print("Building Swin Transformer-inspired Model...")

        def patch_embed(x, patch_size=4, embed_dim=128):
            x = layers.Conv2D(embed_dim, patch_size, strides=patch_size, padding='same')(x)
            return x

        def mlp_block(x, hidden_dim, dropout_rate=0.1):
            original_dim = x.shape[-1]
            x = layers.Dense(hidden_dim, activation='gelu')(x)
            x = layers.Dropout(dropout_rate)(x)
            x = layers.Dense(original_dim)(x)
            x = layers.Dropout(dropout_rate)(x)
            return x

        inputs = layers.Input(shape=(*self.img_size, 3))

        # Patch embedding
        x = patch_embed(inputs, patch_size=4, embed_dim=128)
        x = layers.Reshape((-1, 128))(x)

        # Simplified transformer blocks
        for _ in range(4):
            # Multi-head attention (simplified)
            attn = layers.MultiHeadAttention(num_heads=8, key_dim=16)(x, x)
            x = layers.LayerNormalization()(x + attn)

            # MLP block
            mlp = mlp_block(x, hidden_dim=512)
            x = layers.LayerNormalization()(x + mlp)

        # Global average pooling and classification
        x = layers.GlobalAveragePooling1D()(x)
        x = layers.Dropout(0.5)(x)
        x = layers.Dense(256, activation='gelu')(x)
        x = layers.Dropout(0.3)(x)
        outputs = layers.Dense(self.num_classes, activation='softmax')(x)

        model = models.Model(inputs, outputs)
        return model

    def compile_and_train_model(self, model, model_name, epochs=15):
        """Compile and train a model"""
        print(f"\nTraining {model_name}...")

        # Compile model
        model.compile(
            optimizer=optimizers.Adam(learning_rate=0.001),
            loss='categorical_crossentropy',
            metrics=['accuracy', 'precision', 'recall']
        )

        # Callbacks
        callbacks_list = [
            callbacks.EarlyStopping(
                monitor='val_loss',
                patience=7,
                restore_best_weights=True
            ),
            callbacks.ReduceLROnPlateau(
                monitor='val_loss',
                factor=0.5,
                patience=4,
                min_lr=1e-7
            )
        ]

        # Train model
        history = model.fit(
            self.train_generator,
            epochs=epochs,
            validation_data=self.val_generator,
            class_weight=self.class_weights,
            callbacks=callbacks_list,
            verbose=1
        )

        # Fine-tuning for VGG19
        if model_name == 'VGG19_Transfer':
            print("Fine-tuning VGG19...")
            # Unfreeze some layers for fine-tuning
            model.layers[0].trainable = True
            for layer in model.layers[0].layers[:-4]:
                layer.trainable = False

            # Recompile with lower learning rate
            model.compile(
                optimizer=optimizers.Adam(learning_rate=0.0001),
                loss='categorical_crossentropy',
                metrics=['accuracy', 'precision', 'recall']
            )

            # Continue training
            history_ft = model.fit(
                self.train_generator,
                epochs=8,
                validation_data=self.val_generator,
                class_weight=self.class_weights,
                callbacks=callbacks_list,
                verbose=1
            )

            # Combine histories
            for key in history.history.keys():
                history.history[key].extend(history_ft.history[key])

        self.models[model_name] = model
        self.histories[model_name] = history

        return model, history

    def evaluate_model(self, model, model_name):
        """Evaluate model performance"""
        print(f"\nEvaluating {model_name}...")

        # Predictions on test set
        test_loss, test_acc, test_precision, test_recall = model.evaluate(
            self.test_generator, verbose=0
        )

        # Get predictions for confusion matrix
        self.test_generator.reset()
        predictions = model.predict(self.test_generator, verbose=0)
        predicted_classes = np.argmax(predictions, axis=1)
        true_classes = self.test_generator.classes[:len(predicted_classes)]

        # Calculate metrics
        precision, recall, f1, _ = precision_recall_fscore_support(
            true_classes, predicted_classes, average='weighted'
        )

        # Confusion matrix
        cm = confusion_matrix(true_classes, predicted_classes)

        # Calculate specificity for each class
        specificity = []
        for i in range(self.num_classes):
            tn = np.sum(cm) - (np.sum(cm[i, :]) + np.sum(cm[:, i]) - cm[i, i])
            fp = np.sum(cm[:, i]) - cm[i, i]
            spec = tn / (tn + fp) if (tn + fp) > 0 else 0
            specificity.append(spec)

        avg_specificity = np.mean(specificity)

        results = {
            'test_accuracy': test_acc,
            'test_loss': test_loss,
            'precision': precision,
            'recall': recall,
            'f1_score': f1,
            'specificity': avg_specificity,
            'confusion_matrix': cm,
            'predictions': predictions,
            'true_classes': true_classes,
            'predicted_classes': predicted_classes
        }

        print(f"{model_name} Results:")
        print(f"  Test Accuracy: {test_acc:.4f}")
        print(f"  Test Loss: {test_loss:.4f}")
        print(f"  Precision: {precision:.4f}")
        print(f"  Recall (Sensitivity): {recall:.4f}")
        print(f"  F1-Score: {f1:.4f}")
        print(f"  Specificity: {avg_specificity:.4f}")

        return results

    def plot_training_history(self, model_name):
        """Plot training and validation curves"""
        if model_name not in self.histories:
            print(f"No history found for {model_name}")
            return

        history = self.histories[model_name].history
        epochs = range(1, len(history['accuracy']) + 1)

        fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))
        fig.suptitle(f'{model_name} Training History', fontsize=16)

        # Accuracy
        ax1.plot(epochs, history['accuracy'], 'b-', label='Training Accuracy')
        ax1.plot(epochs, history['val_accuracy'], 'r-', label='Validation Accuracy')
        ax1.set_title('Model Accuracy')
        ax1.set_xlabel('Epoch')
        ax1.set_ylabel('Accuracy')
        ax1.legend()
        ax1.grid(True)

        # Loss
        ax2.plot(epochs, history['loss'], 'b-', label='Training Loss')
        ax2.plot(epochs, history['val_loss'], 'r-', label='Validation Loss')
        ax2.set_title('Model Loss')
        ax2.set_xlabel('Epoch')
        ax2.set_ylabel('Loss')
        ax2.legend()
        ax2.grid(True)

        # Precision
        if 'precision' in history:
            ax3.plot(epochs, history['precision'], 'b-', label='Training Precision')
            ax3.plot(epochs, history['val_precision'], 'r-', label='Validation Precision')
            ax3.set_title('Model Precision')
            ax3.set_xlabel('Epoch')
            ax3.set_ylabel('Precision')
            ax3.legend()
            ax3.grid(True)

        # Recall
        if 'recall' in history:
            ax4.plot(epochs, history['recall'], 'b-', label='Training Recall')
            ax4.plot(epochs, history['val_recall'], 'r-', label='Validation Recall')
            ax4.set_title('Model Recall')
            ax4.set_xlabel('Epoch')
            ax4.set_ylabel('Recall')
            ax4.legend()
            ax4.grid(True)

        plt.tight_layout()
        plt.show()

    def plot_confusion_matrix(self, cm, model_name):
        """Plot confusion matrix"""
        plt.figure(figsize=(10, 8))

        # Create labels for display (remove 'Data_' prefix for cleaner look)
        display_labels = [name.replace('Data_', '').replace('_', ' ').title()
                         for name in self.class_names]

        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                   xticklabels=display_labels,
                   yticklabels=display_labels)
        plt.title(f'{model_name} - Confusion Matrix')
        plt.xlabel('Predicted Label')
        plt.ylabel('True Label')
        plt.tight_layout()
        plt.show()

    def show_sample_predictions(self, model, model_name, num_samples=8):
        """Show sample predictions with images"""
        self.test_generator.reset()

        # Get a batch of test images
        test_batch = next(self.test_generator)
        images, labels = test_batch

        # Make predictions
        predictions = model.predict(images[:num_samples])
        predicted_classes = np.argmax(predictions, axis=1)
        true_classes = np.argmax(labels[:num_samples], axis=1)

        # Plot results
        fig, axes = plt.subplots(2, 4, figsize=(16, 8))
        fig.suptitle(f'{model_name} - Sample Predictions', fontsize=16)

        for i in range(num_samples):
            row = i // 4
            col = i % 4

            axes[row, col].imshow(images[i])
            axes[row, col].axis('off')

            true_label = self.class_names[true_classes[i]].replace('Data_', '')
            pred_label = self.class_names[predicted_classes[i]].replace('Data_', '')
            confidence = predictions[i][predicted_classes[i]]

            color = 'green' if true_classes[i] == predicted_classes[i] else 'red'

            axes[row, col].set_title(f'True: {true_label}\nPred: {pred_label}\nConf: {confidence:.3f}',
                                   color=color)

        plt.tight_layout()
        plt.show()

    def run_complete_analysis(self):
        """Run complete analysis with all models"""
        print("Starting Complete Image Classification Analysis...")
        print("=" * 60)

        # Load data
        self.load_and_preprocess_data()

        # Show sample images from each class
        self.show_sample_images()

        # Build and train all models
        models_to_build = {
            'Custom_CNN': self.build_custom_cnn,
            'VGG19_Transfer': self.build_vgg19_transfer,
            'ConvNeXt_Like': self.build_convnext_model,
            'Swin_Transformer_Like': self.build_swin_transformer_like
        }

        results_summary = {}

        for model_name, build_func in models_to_build.items():
            print(f"\n{'='*60}")
            print(f"Processing {model_name}")
            print(f"{'='*60}")

            try:
                # Build model
                model = build_func()
                print(f"Model architecture built successfully!")
                print(f"Total parameters: {model.count_params():,}")

                # Train model
                model, history = self.compile_and_train_model(model, model_name, epochs=15)

                # Evaluate model
                results = self.evaluate_model(model, model_name)
                results_summary[model_name] = results

                # Plot training history
                self.plot_training_history(model_name)

                # Plot confusion matrix
                self.plot_confusion_matrix(results['confusion_matrix'], model_name)

                # Show sample predictions
                self.show_sample_predictions(model, model_name)

                # Save model
                model.save(f'{model_name}_model.h5')
                print(f"Model saved as {model_name}_model.h5")

            except Exception as e:
                print(f"Error processing {model_name}: {str(e)}")
                import traceback
                traceback.print_exc()
                continue

        # Summary comparison
        self.plot_model_comparison(results_summary)
        self.print_results_table(results_summary)

        return results_summary

    def show_sample_images(self):
        """Display sample images from each class"""
        fig, axes = plt.subplots(1, self.num_classes, figsize=(16, 4))
        fig.suptitle('Sample Images from Each Class', fontsize=16)

        for i, class_name in enumerate(self.class_names):
            class_path = os.path.join(self.data_path, class_name)
            images = [f for f in os.listdir(class_path)
                     if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

            if images:
                img_path = os.path.join(class_path, images[0])
                img = plt.imread(img_path)
                axes[i].imshow(img)
                axes[i].set_title(class_name.replace('Data_', '').replace('_', ' ').title())
                axes[i].axis('off')

        plt.tight_layout()
        plt.show()

    def plot_model_comparison(self, results_summary):
        """Plot comparison of all models"""
        if not results_summary:
            return

        models = list(results_summary.keys())
        metrics = ['test_accuracy', 'precision', 'recall', 'f1_score', 'specificity']

        fig, axes = plt.subplots(1, len(metrics), figsize=(20, 5))

        for i, metric in enumerate(metrics):
            values = [results_summary[model][metric] for model in models]
            bars = axes[i].bar(models, values, color=['blue', 'green', 'red', 'orange'])
            axes[i].set_title(f'{metric.replace("_", " ").title()}')
            axes[i].set_ylabel('Score')
            axes[i].set_ylim(0, 1)

            # Add value labels on bars
            for bar, value in zip(bars, values):
                axes[i].text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
                           f'{value:.3f}', ha='center', va='bottom')

            # Rotate x-axis labels
            axes[i].tick_params(axis='x', rotation=45)

        plt.tight_layout()
        plt.suptitle('Model Performance Comparison', fontsize=16, y=1.02)
        plt.show()

    def print_results_table(self, results_summary):
        """Print results in table format"""
        if not results_summary:
            return

        print("\n" + "="*100)
        print("FINAL RESULTS SUMMARY")
        print("="*100)

        # Create DataFrame for better formatting
        data = []
        for model_name, results in results_summary.items():
            data.append({
                'Model': model_name,
                'Test Accuracy': f"{results['test_accuracy']:.4f}",
                'Precision': f"{results['precision']:.4f}",
                'Recall': f"{results['recall']:.4f}",
                'F1-Score': f"{results['f1_score']:.4f}",
                'Specificity': f"{results['specificity']:.4f}",
                'Test Loss': f"{results['test_loss']:.4f}"
            })

        df = pd.DataFrame(data)
        print(df.to_string(index=False))

        # Find best model for each metric
        print("\n" + "-"*50)
        print("BEST PERFORMING MODELS:")
        print("-"*50)

        metrics = ['test_accuracy', 'precision', 'recall', 'f1_score', 'specificity']
        for metric in metrics:
            best_model = max(results_summary.keys(),
                           key=lambda x: results_summary[x][metric])
            best_score = results_summary[best_model][metric]
            print(f"{metric.replace('_', ' ').title()}: {best_model} ({best_score:.4f})")

# Main execution
if __name__ == "__main__":
    print("Image Classification System")
    print("=" * 60)

    # Extract dataset
    try:
        data_path = extract_dataset()
        print(f"Dataset extracted to: {data_path}")
    except Exception as e:
        print(f"Error extracting dataset: {e}")
        print("Please make sure you upload DataSet.rar file")
        exit()

    # Initialize the system
    classifier = ImageClassificationSystem(
        data_path=data_path,
        img_size=(224, 224),
        batch_size=32
    )

    # Run complete analysis
    print("\nStarting model training and evaluation...")
    results = classifier.run_complete_analysis()

    print("\nAnalysis completed successfully!")
    print("All models have been trained, evaluated, and compared.")
    print("Model files have been saved for future use.")

# Additional utility functions for Colab

def download_best_model(results_summary):
    """Download the best performing model"""
    if not results_summary:
        print("No results available")
        return

    # Find best model by accuracy
    best_model = max(results_summary.keys(),
                    key=lambda x: results_summary[x]['test_accuracy'])

    print(f"Best model: {best_model}")
    print("Downloading model file...")

    try:
        files.download(f'{best_model}_model.h5')
        print("Model downloaded successfully!")
    except:
        print("Error downloading model file")

def create_results_report(results_summary):
    """Create a comprehensive results report"""
    if not results_summary:
        return

    report = "# Image Classification Results Report\n\n"

    for model_name, results in results_summary.items():
        report += f"## {model_name}\n"
        report += f"- **Test Accuracy**: {results['test_accuracy']:.4f}\n"
        report += f"- **Precision**: {results['precision']:.4f}\n"
        report += f"- **Recall**: {results['recall']:.4f}\n"
        report += f"- **F1-Score**: {results['f1_score']:.4f}\n"
        report += f"- **Specificity**: {results['specificity']:.4f}\n"
        report += f"- **Test Loss**: {results['test_loss']:.4f}\n\n"

    # Save report to file
    with open('classification_report.md', 'w') as f:
        f.write(report)

    print("Results report saved as 'classification_report.md'")
    return report

# Optional: Quick test function to verify dataset structure
def verify_dataset_structure(data_path):
    """Verify that the dataset has the correct structure"""
    print("Verifying dataset structure...")

    required_folders = ['Data_cloudy', 'Data_desert', 'Data_green_area', 'Data_water']
    found_folders = []

    if not os.path.exists(data_path):
        print(f" Dataset path not found: {data_path}")
        return False

    for folder in os.listdir(data_path):
        folder_path = os.path.join(data_path, folder)
        if os.path.isdir(folder_path):
            found_folders.append(folder)

            # Count images in folder
            image_files = [f for f in os.listdir(folder_path)
                          if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff'))]
            print(f" {folder}: {len(image_files)} images")

    print(f"\nFound folders: {found_folders}")

    # Check if we have the required structure
    missing_folders = [f for f in required_folders if f not in found_folders]
    if missing_folders:
        print(f"Missing folders: {missing_folders}")
        print("Will use available folders for classification")
    else:
        print("All required folders found!")

    return len(found_folders) > 0

# Enhanced error handling and logging
import logging
import datetime

def setup_logging():
    """Setup logging for the classification system"""
    log_filename = f"classification_log_{datetime.datetime.now().strftime('%Y%m%d_%H%M%S')}.log"

    logging.basicConfig(
        level=logging.INFO,
        format='%(asctime)s - %(levelname)s - %(message)s',
        handlers=[
            logging.FileHandler(log_filename),
            logging.StreamHandler()
        ]
    )

    return logging.getLogger(__name__)

# Performance monitoring
def monitor_training_performance():
    """Monitor GPU/CPU usage during training"""
    try:
        # Check if GPU is available
        if tf.config.list_physical_devices('GPU'):
            print("GPU detected and available for training")
            print("GPU Details:")
            for gpu in tf.config.list_physical_devices('GPU'):
                print(f"  - {gpu}")
        else:
            print("No GPU detected. Training will use CPU (much slower)")

        # Memory info
        print(f"TensorFlow version: {tf.__version__}")
        print(f"Available RAM: {psutil.virtual_memory().available / (1024**3):.1f} GB")

    except Exception as e:
        print(f"Could not get system info: {e}")

# Try to import psutil for system monitoring
try:
    import psutil
    monitor_training_performance()
except ImportError:
    print("Installing psutil for system monitoring...")
    !pip install psutil
    import psutil
    monitor_training_performance()

```
